## 행복 지수(Ladder score)와 6가지 변수들
- **행복지수에 영향을 줄 수 있는 6가지 변수들**에 대하여 어떤 양상을 보이는지 확인
- 행복지수와 관련이 높은 변수에는 1인 국내총생산, 사회적 지지도, 기대 수명이 있다.

```python
variable = ['Ladder score', 'Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
happiness_df[variable].corr()
```

```python
sns.heatmap(happiness_df[variable].corr())
```

### 행복 지수 vs  1인당 국내총생산
- GDP가 높아질수록 행복지수도 높아진다.

```python
plt.xlabel('Logged GDP per capita')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Logged GDP per capita'],happiness_df['Ladder score'])
plt.show()
```

### 행복 지수 vs 사회적 지지도
- 사회적 지지도가 높아질수록 행복지수도 높아진다.

```python
plt.xlabel('Social support')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Social support'],happiness_df['Ladder score'])
plt.show()
```

### 행복 지수 vs 기대 수명
- 기대 수명이 높아질수록 행복지수도 높아진다.

```python
plt.xlabel('Healthy life expectancy')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Healthy life expectancy'],happiness_df['Ladder score'])
plt.show()
```

### 행복 지수 vs 선택의 자유
- 2021년도의 선택의 자유는 대부분 0.6을 넘고, 높은 편이다. 하지만, 선택의 자유가 높아도 반드시 행복 지수가 높은 것이 아니다.

```python
plt.xlabel('Freedom to make life choices')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Freedom to make life choices'],happiness_df['Ladder score'])
plt.show()
```

### 행복 지수 vs 관용
- 기부를 많이 할수록 행복 지수가 높을 줄 알았는데 오히려 대부분 0이하의 값이 나왔다. 
- 행복 지수와 관계 없이 대부분의 나라의 기부 수준은 낮았다.
- 오히려 관용 점수가 가장 높은 4개의 나라의 행복 지수는 6점 아래다.

```python
plt.xlabel('Generosity')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Generosity'],happiness_df['Ladder score'])
plt.show()
```

### 행복 지수 vs 부정부패 인식
- 부정부패 인식의 정도가 높아질수록 행복지수도 낮아졌지만 정도가 심하지 않았다. 
- 오히려 대부분의 나라들이 부정부패 인식 수준이 매우 높았고, 그에 따른 행복지수는 4~7에 몰려있다. 

```python
plt.xlabel('Perceptions of corruption')
plt.ylabel('Ladder score')
plt.scatter(happiness_df['Perceptions of corruption'],happiness_df['Ladder score'])
plt.show()
```

### 지역별 행복 지수
- 대체로 GDP가 높은 지역일수록 행복 지수가 높다.

```python
Region_df = happiness_df[['Ladder score','Regional indicator','Logged GDP per capita']].groupby('Regional indicator').median().iloc[:,:]
Region_df = Region_df.sort_values(by='Ladder score' ,ascending=True)
Region_df
```

```python
bar_width = 0.2
x = np.arange(len(Region_df.index))
plt.barh(x-0.0, Region_df.iloc[:,0],bar_width,label = 'Ladder score')
plt.barh(x+0.2, Region_df.iloc[:,1],bar_width,label = 'GDP')
plt.yticks(x,Region_df.index)
plt.legend(loc = 'center left', bbox_to_anchor=(1,0.5))
plt.show()
```
